```python
from pdfminer.pdfpage import PDFPage
from pdfminer.pdfinterp import PDFResourceManager
from pdfminer.pdfinterp import PDFPageInterpreter
from pdfminer.layout import LAParams, LTTextBox, LTTextLine, LTFigure, LTImage, LTChar
from pdfminer.converter import PDFPageAggregator

import pprint as pp
    
output=open('sample.txt', 'w')

#Open a PDF file.
fp = open('FOMC20080130meeting.pdf', 'rb')

#Create a PDF resource manager object that stores shared resources.
rsrcmgr = PDFResourceManager()

# Set parameters for analysis.
laparams = LAParams()

# Create a PDF device object.
device = PDFPageAggregator(rsrcmgr, laparams=laparams)

# Create a PDF interpreter object.
pdf_interp = PDFPageInterpreter(rsrcmgr, device)

pagedata=[]

for number, page in enumerate(PDFPage.get_pages(fp, password='')):
    pdf_interp.process_page(page)
    layout = device.get_result()
    text_content=[]
    for obj in layout:
        #print obj
        if isinstance(obj, LTTextBox) or isinstance(obj, LTTextLine):
            text_content.append(obj.get_text().strip())
        else :
            pass
    pagedata.append(text_content)

for pg in pagedata:
    output.write('\n'.join(pg).encode('utf-8'))
    output.write('\n'.encode('utf-8'))

fp.close()                 
output.close()
```

```python
import nltk

fed_jan=open('sample.txt')

tokens = nltk.word_tokenize(fed_jan.read())
text1 = nltk.Text(tokens)

fed_jan.close()
```

```python
from nltk.probability import FreqDist

def removeNonAscii(s): 
    return "".join(filter(lambda x: ord(x)<128, s))

fdist1 = FreqDist(text1)
fdist1.plot(50, cumulative=False)

rare_words = fdist1.hapaxes()
rare_words_pretty=[removeNonAscii(w) for w in rare_words]
print ('Number of rare words: %d' % len(rare_words_pretty))
rare_words_pretty
```
